```python
# Convert order_date to datetime 
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
# Group sales by month and year
df['YearMonth'] = df['order_date'].dt.to_period('M')
sales_over_time = df.groupby('YearMonth')['total_amount'].sum().reset_index()
```

```python
# Convert period to string for labeling
sales_over_time['YearMonth'] = sales_over_time['YearMonth'].astype(str)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(sales_over_time['YearMonth'], sales_over_time['total_amount'], marker='o', color='blue', linewidth=2)
plt.title("Sales Trend (by Month & Year)"),
plt.xlabel("Month & Year")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.show()
```

Sales show fluctuating growth across 2023â€“2025 with seasonal spikes around December (holiday period).

Dip observed in september 2013, 2024,and also 2025 possibly due to off-peak demand.

Insight: Sales seasonality exists â€” likely driven by festive or promotional events.

Recommendation:

Plan targeted marketing campaigns around high-performing months.

Offer discounts or free shipping during low-demand months to balance revenue.

```python
Sales_by_region= df.groupby("region")["total_amount"].sum().reset_index().sort_values(by="total_amount",ascending =False)
```

```python
Sales_by_region
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot( data=Sales_by_region, x="region", y="total_amount",hue="region",palette="Set2")
plt.title("Price Distribution By Region")
plt.xlabel("region")
plt.ylabel("Amount")
plt.xticks(rotation=15)
plt.show()
```

South and North regions lead in total sales.

West and East shows moderate activity, while Central is less engaged.

Insight: Regional demand varies significantly.

Recommendation:

Focus ads and delivery optimization in South/North.

Conduct surveys in low-performing regions to identify barriers (e.g., shipping cost, awareness).


```python
#category
Top_5_product_category_by_sales = df.groupby("category")["total_amount"].sum().reset_index().sort_values(by="total_amount",ascending =False)[:5]
```

```python
Top_5_product_category_by_sales
```

```python
plt.figure(figsize=(8,5))
sns.barplot( data=Top_5_product_category_by_sales, x="category", y="total_amount",hue="category",palette="Set2")
plt.title("Top 5 Product Category by Revenue")
plt.xlabel("product category")
plt.ylabel("Amount")
plt.xticks(rotation=15)
plt.show()
```

Top categories: Electronics,Home and sports are the top 3 product categories that contribute the most to total revenue.

Least performing: Fashion and Beauty.

Insight: Few categories dominate revenue.

Recommendation:

Strengthen inventory and promotion for Fashion,Beauty and other categories that are least performing.

Run promotional offers and campaigns for underperforming categories like Beauty.
